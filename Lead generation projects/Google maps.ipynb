```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time

def scrape_google_maps():
    # Configure Chrome options for incognito mode
    chrome_options = Options()
    chrome_options.add_argument("--incognito")
    chrome_options.add_argument("--start-maximized")

    # Initialize WebDriver
    driver = webdriver.Chrome(options=chrome_options)

    try:
        # Step 1: Open Google Maps and search for the query
        driver.get("https://www.google.com/maps")
        time.sleep(5)  # Allow the page to load fully

        search_box = driver.find_element(By.ID, "searchboxinput")
        search_box.send_keys("graphic design institutes near vellore, Tamil Nadu")
        search_box.send_keys(Keys.RETURN)
        time.sleep(5)  # Wait for search results to load

        # Initialize lists to store data
        headings = []
        phone_numbers = []

        # Infinite scrolling to load all results
        scroll_pause_time = 3
        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
            # Extract headings
            heading_elements = driver.find_elements(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]/div[3]/div/a')
            for heading_element in heading_elements:
                try:
                    heading = heading_element.text
                    if heading not in headings:  # Avoid duplicates
                        headings.append(heading)
                        print(f"Heading: {heading}")
                except Exception as e:
                    print(f"Error extracting heading: {e}")

            # Extract phone numbers
            phone_elements = driver.find_elements(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[7]/div[5]/button/div/div[2]/div[1]')
            for phone_element in phone_elements:
                try:
                    phone = phone_element.text
                    if phone not in phone_numbers:  # Avoid duplicates
                        phone_numbers.append(phone)
                        print(f"Phone Number: {phone}")
                except Exception as e:
                    print(f"Error extracting phone number: {e}")

            # Scroll down to load more content
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(scroll_pause_time)

            # Check if more content is loaded
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                print("No more content to load.")
                break
            last_height = new_height

        # Print final results
        print("\nFinal Results:")
        print(f"Headings: {headings}")
        print(f"Phone Numbers: {phone_numbers}")

    finally:
        # Close the browser
        driver.quit()

if __name__ == "__main__":
    scrape_google_maps()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# Initialize WebDriver
driver = webdriver.Chrome()  # Ensure ChromeDriver is in your PATH
driver.get("https://www.google.com/maps")

def search_google_maps(query, scroll_pause=5):
    """
    Searches Google Maps and extracts business details until all results are loaded.
    Args:
        query (str): The search term (e.g., "restaurants near me").
        scroll_pause (int): Time to pause while scrolling, in seconds.
    Returns:
        list: A list of dictionaries with business details.
    """
    # Input search query
    search_box = driver.find_element(By.ID, "searchboxinput")
    search_box.clear()
    search_box.send_keys(query)
    search_box.send_keys(Keys.ENTER)
    time.sleep(scroll_pause)  # Wait for results to load

    business_details = []
    last_result_count = 0  # To track the number of results

    while True:
        # Find all result cards
        result_cards = driver.find_elements(By.CLASS_NAME, "Nv2PK")
        
        # Check if new results are loaded
        if len(result_cards) == last_result_count:
            print("All results loaded.")
            break

        last_result_count = len(result_cards)

        # Extract details from each card
        for card in result_cards[len(business_details):]:  # Avoid re-scraping already scraped results
            try:
                name = card.find_element(By.CLASS_NAME, "qBF1Pd").text
                address = card.find_element(By.CLASS_NAME, "W4Efsd").text
                try:
                    rating = card.find_element(By.CLASS_NAME, "MW4etd").text
                except:
                    rating = "No rating"
            except Exception as e:
                print(f"Skipping a card due to missing information: {e}")
                continue

            business_details.append({
                "Name": name,
                "Address": address,
                "Rating": rating
            })

        # Scroll down to load more results
        driver.execute_script("arguments[0].scrollIntoView();", result_cards[-1])
        time.sleep(scroll_pause)  # Wait for more results to load

    return business_details

# Use the function to fetch businesses
query = "cafes in cp"
businesses = search_google_maps(query, scroll_pause=7)

# Use the function to fetch businesses
query = "graphic design institute kochi"
zoom_out_map(times=2, pause=3)  # Adjust the zoom level
businesses = search_google_maps(query, scroll_pause=7)

# Save to a DataFrame and export as CSV
try:
    df = pd.DataFrame(businesses)
    df.to_csv("businesses.csv", index=False)
    print("Details saved to businesses.csv")
except PermissionError:
    print("Permission denied: Please ensure 'businesses.csv' is not open in another program.")
finally:
    driver.quit()
```
